```python
import sqlite3
import pandas
import pymongo
client = pymongo.MongoClient()
db = client.sae

# Création de la connexion
conn = sqlite3.connect("Gymnase2000.sqlite")
```

```python
gymnases = pandas.read_sql_query(
    """SELECT *
    FROM Gymnases;
    """, 
    conn
)
```

```python
seances = pandas.read_sql_query(
    """
    SELECT *
    FROM Seances
    INNER JOIN Sports
    USING (IdSport);
    """,
    conn
)
```

```python
sessions = [
    seances.query('IdGymnase == @id')
        .drop(columns=["IdGymnase", "IdSport"])
        .to_dict(orient = "records") 
        for id in gymnases.IdGymnase
]
print(sessions)
```

```python
gymnases = gymnases.assign(Sessions = sessions)
gymnases.head()
```

```python
db.gymnases.insert_many(
    gymnases.to_dict(orient = "records")
)
```

```python
Sportifs = pandas.read_sql_query ("""select * from Sportifs""", conn) 
Arbitrer = pandas.read_sql_query ("""select * from Arbitrer""", conn) 
Jouer = pandas.read_sql_query ("""select * from Jouer""", conn) 
Entrainer = pandas.read_sql_query ("""select * from Entrainer""", conn) 
Sports=pandas.read_sql_query ("""select * from Sports""", conn) 
```

```python
id = 7
sport7 = Sportifs.query('IdSportif == @id')
print(sport7)
```

```python
# Ajouter les activités à chaque sportif
Sportifs["Arbitrer"] = Sportifs["IdSportif"].apply(
    lambda id: Arbitrer[Arbitrer["IdSportif"] == id]["IdSport"].tolist()
)
Sportifs["Jouer"] = Sportifs["IdSportif"].apply(
    lambda id: Jouer[Jouer["IdSportif"] == id]["IdSport"].tolist()
)
Sportifs["Entrainer"] = Sportifs["IdSportif"].apply(
    lambda id: Entrainer[Entrainer["IdSportifEntraineur"] == id]["IdSport"].tolist()
)
# Vérifier la structure des données combinées
print(Sportifs.head())
```

```python
# Créer un dictionnaire pour mapper les IDs des sports à leurs libellés
sports_dict = Sports.set_index("IdSport")["Libelle"].to_dict()

# Ajouter les libellés des sports pour chaque activité
Arbitrer["Libelle"] = Arbitrer["IdSport"].map(sports_dict)
Jouer["Libelle"] = Jouer["IdSport"].map(sports_dict)
Entrainer["Libelle"] = Entrainer["IdSport"].map(sports_dict)
```

```python
# Ajouter les libellés des sports arbitrés par chaque sportif
Sportifs["Arbitrer"] = Sportifs["IdSportif"].apply(
    lambda id: Arbitrer[Arbitrer["IdSportif"] == id]["Libelle"].tolist()
)

# Ajouter les libellés des sports joués par chaque sportif
Sportifs["Jouer"] = Sportifs["IdSportif"].apply(
    lambda id: Jouer[Jouer["IdSportif"] == id]["Libelle"].tolist()
)

# Ajouter les libellés des sports entraînés par chaque sportif
Sportifs["Entrainer"] = Sportifs["IdSportif"].apply(
    lambda id: Entrainer[Entrainer["IdSportifEntraineur"] == id]["Libelle"].tolist()
)

# Vérifier le résultat
print(Sportifs.head())
```

```python
db.Sportifs.insert_many(
    Sportifs.to_dict(orient = "records")
)
```

```python
db.Sportifs.count_documents({})
```

```python
pandas.DataFrame(list(db.gymnases.aggregate([
    { "$limit": 1 },
    { "$unwind": "$Sessions" },
    { "$lookup": {
       "from": "Sportifs",
       "localField": "Sessions.IdSportifEntraineur",
       "foreignField": "IdSportif",
       "as": "Entraineur"
     }}
])))
```

```python
import pandas as pd
# Charger les données des sportifs avec les informations des conseillers
sportifs_with_conseillers = list(db.Sportifs.aggregate([
    {
        "$lookup": {
            "from": "sportifs",  # Jointure avec la même collection
            "localField": "IdSportifConseiller",  # Champ à utiliser pour la jointure
            "foreignField": "IdSportif",  # Champ dans la collection "sportifs"
            "as": "Conseiller"  # Nom du champ pour le résultat
        }
    }
]))

# Convertir le résultat en DataFrame pour l'affichage
df_sportifs = pd.DataFrame(sportifs_with_conseillers)
print(df_sportifs.head())
```
